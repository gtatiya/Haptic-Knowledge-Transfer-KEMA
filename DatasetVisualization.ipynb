# Discretized Data

```python
path = r"Datasets"

#db_file_name = "baxter_shake_discretizedmean-10.bin"
#db_file_name = "baxter_shake_discretizedrange-15.bin"

#db_file_name = "fetch_shake_discretizedmean-10.bin"
#db_file_name = "fetch_shake_discretizedrange-15.bin"

#db_file_name = "sawyer_shake_discretizedmean-10.bin"
db_file_name = "sawyer_shake_discretizedrange-15.bin"
```

```python
def read_dataset_discretized(a_path, db_file_name, temporal_bins):

    bin_file = open(a_path+os.sep+db_file_name, "rb")

    examples = pickle.load(bin_file)
    labels = pickle.load(bin_file)
    bin_file.close()
    
    return examples.reshape(NUM_OF_OBJECTS, TRIALS_PER_OBJECT, CHANNELS, temporal_bins, -1), labels.reshape(NUM_OF_OBJECTS, TRIALS_PER_OBJECT)
    
    
def plot_an_example_discretized(an_example, from_to, modality, robot, behavior, dataset_type, save_plot):
    
    modality = modality.split(' ')[1]
    print("Feature Extraction Method: ", dataset_type)
    title = robot+"-"+behavior+"-"+modality+" Features"
    
    plt.title(title, fontsize=16)
    plt.xlabel('Temporal Bins', fontsize=16)
    plt.ylabel('Joints', fontsize=16)

    plt.imshow(an_example[:, from_to[0]:from_to[1]].T, cmap="gist_ncar")
    
    x_values, y_values = an_example[:, from_to[0]:from_to[1]].shape
    ax = plt.gca()
    ax.set_xticks(np.arange(0, x_values, 1))
    ax.set_yticks(np.arange(0, y_values, 1))
    ax.set_xticklabels(np.arange(1, x_values + 1, 1))
    ax.set_yticklabels(np.arange(1, y_values + 1, 1))
    
    plt.colorbar(orientation='vertical')
    
    if save_plot:
        file_path = "figs"+os.sep+robot+"_"+behavior+"_"+modality+"_"+dataset_type+".png"
        plt.savefig(file_path, bbox_inches='tight', dpi=100)

    plt.show()
    plt.close()
```

```python

robot_name, behavior_name, dataset_type = db_file_name.split('_')[0].capitalize(), db_file_name.split('_')[1].capitalize(), db_file_name.split('_')[2][:-4].capitalize()
temporal_bins = int(dataset_type.split('-')[1])

examples, labels = read_dataset_discretized(path, db_file_name, temporal_bins)

print("Robot: ", robot_name)
print("Behavior: ", behavior_name)
print("Dataset Type: ", dataset_type)
print("Temporal Bins: ", temporal_bins)

trial_no = 0
obj_no = 11
channel = 0 #0 for mean

print("Example:")
print("Objects, Trials, Channels, Temporal Bins, Features", examples.shape)
print(examples[obj_no][trial_no][channel].shape)

print("\nLabel:")
print(labels.shape)
```

```python
plot_an_example_discretized(examples[obj_no][trial_no][0], [0, 9], "Dediscretized Position Mean", robot_name, behavior_name, dataset_type, False)

plot_an_example_discretized(examples[obj_no][trial_no][0], [9, 18], "Dediscretized Velocity Mean", robot_name, behavior_name, dataset_type, False)

plot_an_example_discretized(examples[obj_no][trial_no][0], [18, 27], "Dediscretized Effort Mean", robot_name, behavior_name, dataset_type, True)
```
